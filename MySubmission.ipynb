```python
df = pd.read_csv("training_set_features.csv")
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
for x in df.columns:
    if df[x].dtype==np.float64:
        df[x] = df[x].fillna(df[x].mode()[0])
```

```python
merged = df
l = []
for x in df.columns:
    if df[x].dtype==object:
        l.append(x)
        merged = pd.concat([merged, pd.get_dummies(df[x])],axis = 'columns')
merged
```

```python
merged.drop(l,axis = 1,inplace = True)
merged
```

```python
merged.fillna(0,inplace= True)
```

```python
merged.isna().sum().sum() #checking
```

```python
merged.drop('respondent_id',axis = 1,inplace = True)
```

```python
y = pd.read_csv('training_set_labels.csv')
y.drop('respondent_id',axis = 1, inplace = True)
```

```python
y.columns
```

```python
merged.head()
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline
```

```python
steps = [('std',StandardScaler()),('lr',LogisticRegression())]
```

```python
pipe = Pipeline(steps)
```

```python
from sklearn import set_config
set_config(display = "diagram")
```

```python
pipe
```

```python
pipe.fit(merged,y.xyz_vaccine)
```

```python
tester = pd.read_csv("test_set_features.csv")
for x in tester.columns:
    if tester[x].dtype==np.float64:
        tester[x] = tester[x].fillna(tester[x].mode()[0])
```

```python
x_test = tester
l = []
for x in tester.columns:
    if tester[x].dtype==object:
        l.append(x)
        x_test = pd.concat([x_test, pd.get_dummies(tester[x])],axis = 'columns')
x_test
```

```python
x_test.drop(l,axis = 1,inplace = True)
x_test.fillna(0,inplace= True)
```
